```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.3)
```

```python
sns.scatterplot(X_train[:,0],X_train[:,1], hue=y_train)
```

```python
model1.add(dense(1,input_dim=2, activation='sigmoid'))
```

```python
model1.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
model1.fit(X_train,y_train,epochs=100
          )
```

```python
y_pred=model1.predict(X_test)
```

```python
from sklearn.metrics import classification_report as cvs, confusion_matrix as cm
```

```python
y_pred
```

```python
y_pred=[1 if y_pred[i]>0.5 else 0 for i in range(len(y_pred))]
```

```python
sns.heatmap(cm(y_test,y_pred), annot=True ,cmap='viridis')
```

```python
report=cvs(y_test,y_pred)
```

```python
print(report)
```
